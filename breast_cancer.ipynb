## Convert to Pandas

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sn

dataset = df.toPandas()
```

## Display more info

```python
dataset.head()
```

```python
dataset.info()
```

## Remove "id" and "_c32" columns

```python
dataset = dataset.drop(["id"], axis = 1)
dataset = dataset.drop(["_c32"], axis = 1)
```

```python
dataset.head(3)
```

```python
list_of_columns_without_diagnosis = list(dataset.columns)
list_of_columns_without_diagnosis.pop(0)
dataset[list_of_columns_without_diagnosis] = dataset[list_of_columns_without_diagnosis].apply(pd.to_numeric)
dataset.info()
```

## Dataset Visualization

```python
M = dataset[dataset.diagnosis == "M"]
```

```python
M.head(5)
```

```python
B = dataset[dataset.diagnosis == "B"]
```

```python
B.head(5)
```

```python
dataset.diagnosis = [1 if i == "M" else 0 for i in dataset.diagnosis]
```

```python
corr_matrix = dataset.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
cmap = sn.diverging_palette(220, 20, as_cmap=True)
plt.figure(figsize=(20,20))
sn.heatmap(corr_matrix, annot=True, mask=mask, cmap=cmap)
plt.show()
```

## Decision Tree with Sklearn

```python
from sklearn.model_selection import train_test_split

x = dataset.drop(["diagnosis"], axis = 1)
y = dataset.diagnosis.values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=3)
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
from sklearn.tree import plot_tree

feature_names = list(dataset.columns)
feature_names.pop(0)
class_names = ["B", "M"]

plt.figure(figsize=(30,8))
plot_tree(decision_tree=dt, feature_names=feature_names, class_names=class_names)
plt.show()
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

classif_report = classification_report(y_test, y_pred)
print(classif_report)

cm = confusion_matrix(y_test, y_pred)
print(cm)
cm_classes = ["B", "M"]
cm_array_df = pd.DataFrame(cm, index=cm_classes, columns=cm_classes)
cm_cmap = sn.diverging_palette(20, 220, as_cmap=True)
sn.heatmap(cm_array_df, annot=True, cmap=cm_cmap, cbar=False, fmt='g')
plt.xlabel('Predito')
plt.ylabel('Real')
```
